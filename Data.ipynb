```python
import pandas as pd
import seaborn as sns
sns.set(color_codes = True)
from matplotlib import pyplot as plt
```

```python
health_set = pd.read_csv("Health.csv")
health_set.head(10)
```

```python
health_set.info()
```

```python
type(health_set)
```

```python
sns.lineplot(x = 'Year', y = 'Value', hue = 'Measure', data =health_set)
Year= health_set['Year']
plt.title("Tuberculosis Deaths")
plt.xticks(health_set['Year'])
plt.rcParams["figure.figsize"] = (22,3)
```

```python
sns.pairplot( hue = 'Measure', data =health_set)
plt.rcParams["figure.figsize"] = (22,3)
plt.show()
```

```python
sns.barplot (x = 'Year', y = 'Value', hue = 'Measure', data =health_set)
plt.rcParams["figure.figsize"] = (22,3)
```

```python
sns.pointplot (x = 'Year', y = 'Value', hue = 'Measure', data =health_set,  markers=["o", "*", "^"])
plt.rcParams["figure.figsize"] = (22,3)
```

```python
sns.lmplot (x = 'Year', y = 'Value', hue = 'Measure', data =health_set,  markers=["o", "*", "^"])
plt.rcParams["figure.figsize"] = (22,3)
```
