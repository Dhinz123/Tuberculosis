```python
import pandas as pd
import seaborn as sns
sns.set(color_codes = True)
from matplotlib import pyplot as plt
```

```python
health_set = pd.read_csv("Health.csv")
health_set.head(10)
```

```python
health_set.info()
```

```python
type(health_set)
```

```python
sns.lineplot(x = 'Year', y = 'Value', hue = 'Measure', data =health_set)
plt.title("Tuberculosis Deaths")
```

```python
sns.boxplot(x = 'Year', y = 'Value', hue = 'Measure', data =health_set)
```

```python
sns.barplot (x = 'Year', y = 'Value', hue = 'Measure', data =health_set)
```

```python
sns.pointplot (x = 'Year', y = 'Value', hue = 'Measure', data =health_set)
```
